```python
import os
import csv
import re
dir_path=os.path.expanduser("~\DataMiner\hs_data")
dir_path=os.path.expanduser("~/DataMiner/hs_data")
all_text=[]
names=[]

links=[]
heads=[]
texts=[]
all_texts=[]

print(dir_path)
for file in sorted(os.listdir(dir_path),reverse=True):
    if file.endswith("csv"):
        print(file)
        #file="2019-03-01 hs data.csv"
        with open(os.path.join(dir_path,file)) as csvfile:
            day_links=[]
            day_heads=[]
            day_texts=[]
            names.append(file)
            reader=csv.reader(csvfile, delimiter=';')
            for row in reader:
                try:
                    if "interactive" not in row[0]:
                        try:
                            day_links.append(row[0])
                            day_heads.append(row[1])
                            day_texts.append(row[2].lower())
                            all_texts.append(row[2].lower())
                        except KeyboardInterrupt:
                            raise
                        except IndexError:
                            print("Row is ", row)
                            print(row[1])
                            if len(day_links)>len(all_texts):
                                print("Removing ",day_links.pop())
                            if len(day_heads)>len(all_texts):
                                print("Removing ",day_heads.pop())
                            if len(day_texts)>len(day_texts):
                                print("Removing ",day_links.pop())
                                
                except KeyboardInterrupt:
                    raise
                #except:
                #    print("Next val")
            links.append(day_links)
            heads.append(day_heads)
            texts.append(day_texts)
```

```python
import csv, os
def writeToCsv(file,data,writeappend):
    file=os.path.expanduser(file)
    with open(file,writeappend) as f:
        wr = csv.writer(f, dialect='excel')
        wr.writerows(data)
        
def readFromCsv(file):
    file=os.path.expanduser(file)
    with open(file,'r') as f:
        reader = csv.reader(f, dialect='excel')
        data=[]
        for r in reader:
            data.append(r)
    return data

import os, psutil, numpy as np
def usage():
    process = psutil.Process(os.getpid())
    return process.memory_info()[0] / float(2 ** 20)

unranked=None
path=os.path.expanduser("~/DataMiner/unranked.csv")

try:
    if unranked==None: raise
    writeToCsv(path,unranked,'w')
    print("Written to csv")
except:
    unranked=readFromCsv(path)
    '''for row in unranked:
        print(row)'''
print("Done")
```
